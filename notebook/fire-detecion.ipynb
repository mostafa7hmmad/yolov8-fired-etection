```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')
model.train(
    data='/kaggle/input/model-yolo/other/default/1/Data.yaml',
    epochs=30,
    batch=32,  # ØªÙ‚Ù„ÙŠÙ„ Ø­Ø¬Ù… Ø§Ù„Ø¯ÙØ¹Ø©
    imgsz=640,
    optimizer='Adam',
    lr0=0.0005,  # ØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù…
    weight_decay=0.001,  # Ø²ÙŠØ§Ø¯Ø© ÙˆØ²Ù† Ø§Ù„ØªØµØ­ÙŠØ­
    patience=7,
    save=True,
    cos_lr=True,
    mosaic=0.8,
    fliplr=0.5,
    degrees=10.0,  # Ø²ÙŠØ§Ø¯Ø© Ø§Ù„ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ø²Ø§ÙˆÙŠ
    translate=0.2,  # Ø²ÙŠØ§Ø¯Ø© Ø§Ù„ØªØ±Ø¬Ù…Ø©
    scale=0.7,  # ØªØºÙŠÙŠØ± Ø§Ù„Ø­Ø¬Ù…
    shear=0.3,  # Ø¥Ø¶Ø§ÙØ© Ø¥Ù…Ø§Ù„Ø©
    verbose=True,
    plots=True
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Path to the results.csv file in train15
results_path = r"detect\train2\results.csv"

# Check if the file exists
if not os.path.exists(results_path):
    print(f"results.csv not found at {results_path}. Please check the path or folder.")
    exit()

# Load the CSV file
df = pd.read_csv(results_path)

# Strip any whitespace from column names (YOLOv8 sometimes adds spaces)
df.columns = df.columns.str.strip()

# Plot 1: Box Loss
plt.figure(figsize=(10, 5))
plt.plot(df["epoch"], df["train/box_loss"], label="Train Box Loss", color="blue")
plt.plot(df["epoch"], df["val/box_loss"], label="Val Box Loss", color="orange")
plt.xlabel("Epoch")
plt.ylabel("Box Loss")
plt.title("Training and Validation Box Loss (train15)")
plt.legend()
plt.grid()
plt.show()

# Plot 2: Classification Loss
plt.figure(figsize=(10, 5))
plt.plot(df["epoch"], df["train/cls_loss"], label="Train Cls Loss", color="blue")
plt.plot(df["epoch"], df["val/cls_loss"], label="Val Cls Loss", color="orange")
plt.xlabel("Epoch")
plt.ylabel("Cls Loss")
plt.title("Training and Validation Classification Loss (train15)")
plt.legend()
plt.grid()
plt.show()

# Plot 3: DFL Loss (if available)
if "train/dfl_loss" in df.columns and "val/dfl_loss" in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df["epoch"], df["train/dfl_loss"], label="Train DFL Loss", color="blue")
    plt.plot(df["epoch"], df["val/dfl_loss"], label="Val DFL Loss", color="orange")
    plt.xlabel("Epoch")
    plt.ylabel("DFL Loss")
    plt.title("Training and Validation DFL Loss (train15)")
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("DFL Loss columns not found in results.csv. Skipping DFL Loss plot.")
```
